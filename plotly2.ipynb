```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
matches=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
# matches['batsman']
# matches.shape
```

```python
matches.head()
```

```python
delivery.head()
# delivery.shape
# delivery['batsman']
```

```python
ipl=delivery.merge(matches,left_on='match_id',right_on='id')
# ipl['batsman_runs']
```

```python
top50 = ipl.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(50)
# top50
new_ipl=ipl[ipl['batter'].isin(top50.index)]
new_ipl
```

```python
runs=new_ipl.groupby('batter')['batsman_runs'].sum()
balls=new_ipl.groupby('batter')['batsman_runs'].count()

sr=(runs/balls)*100
sr=sr.reset_index()
sr
```

```python
ot=ipl[ipl['player_dismissed'].isin(top50.index)]
nots=ot['player_dismissed'].value_counts()
avg=runs/nots
avg=avg.reset_index()
avg.rename(columns={'index':'batter',0:'avg'},inplace=True)
avg=avg.merge(sr,on='batter')
avg
```

```python
trace=go.Scatter(x=avg['avg'],y=avg['batsman_runs'],mode='markers')

data=[trace]
layout=go.Layout(title='Avg vs SR',xaxis={'title':'Avg'},yaxis={'title':'SR'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
fig.show()
```

```python
single=ipl[ipl['batter']=='MS Dhoni']
single
perf=single.groupby('season')['batsman_runs'].sum().reset_index()
perf
```

```python
trace=go.Scatter(x=perf['season'],y=perf['batsman_runs'],mode='lines')
data=[trace]
layout=go.Layout(title='MS Dhoni',xaxis={'title':'season'},yaxis={'title':'runs'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
fig.show()
```

```python

top10 = ipl.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top10_df=ipl[ipl['batter'].isin(top10.index)]
top10_df
```

```python
total_score=top10_df.groupby(['batter'])['batsman_runs'].sum().reset_index()
total_score
```

```python
tarce=go.Bar(x=total_score['batter'],y=total_score['batsman_runs'])
data=[tarce]
layout=go.Layout(title='Top 10 Batsman',xaxis={'title':'Batsman'},yaxis={'title':'Total Runs'})
fig=go.Figure(data=data,layout=layout)
pyo.plot(fig)
fig.show()
```
